#### Code & Test

```python
# I took all of the dogs that are not puppo, pupper, floofer, doggo so we can merge them afterward
non_type = df1[df1['doggo']=='None']
non_type = non_type[non_type['floofer'] == 'None']
non_type = non_type[non_type['pupper'] == 'None']
non_type = non_type[non_type['puppo'] == 'None']
```

```python
non_type.head(25)
```

```python
df1_copy = df1.copy() # I took a copy from the original df1
```

```python
df1_copy = pd.melt(df1_copy , id_vars=['tweet_id','timestamp','rate','name'] , var_name='type') # then melt all the dog types (puppo, pupper, floofer, doggo)
```

```python
df1_copy = df1_copy[df1_copy.value != 'None'] # drop all the rows with type value = None
```

```python
df1_copy.head()
```

```python
df1_copy = df1_copy.drop(columns=['value']) # droping the value column since it's useless
```

```python
df1_copy.shape[0] + non_type.shape[0] == df1.shape[0] # it seems that there are some duplicates
```

```python
df1_copy.tweet_id.duplicated().sum() # it turned out that there are some dogs with multible types 
```

```python
df1_copy.loc[df1_copy.tweet_id.duplicated(keep=False)] # we are going to take a look at the duplicated rows
# so we can see that there are 14 dogs with two types and since by definition from the (THE DOGTIONARY) url:(https://classroom.udacity.com/nanodegrees/nd002-connect/parts/f3fb7339-2000-47dd-a21e-29480cdd8166/modules/14d9f5f1-9e7b-4bfb-97f3-bcdbf4a3699c/lessons/a8085857-3e28-4fc7-aeb8-da64ccbc2e20/concepts/28d4643b-3785-4700-bdee-4e5fc9963576)
# A dog can be only 1 type based on how he described them in his book, so I'm just going to assume that a dog can only be 1 type and I'll remove the duplicated ones.
```

```python
df1_copy.drop_duplicates(subset='tweet_id',inplace=True) # so we are going to drop all duplicated tweet id
```

```python
df1_copy.shape[0] + non_type.shape[0] == df1.shape[0]
```

# Analyzing and Visualizing

```python
# First insights 
mrd = WeRateDogs.loc[WeRateDogs.retweet_count.idxmax()] # Most Retweeted Dog
mrd # The most retweeted dog is a Labrador retriever dog that we don't know this name
```

```python
# second insights 
mfd = WeRateDogs.loc[WeRateDogs.favorite_count.idxmax()] # Most favorited Dog
mfd # The most favorited dog is a Lakeland terrier dog that we don't know this name
```

```python
# 3rd insights 
higest_rate = WeRateDogs.loc[WeRateDogs.rate.idxmax()] # Most favorited Dog
higest_rate # The highest rated dog is a Pomeranian dog by the name Logan
```

```python
viz = WeRateDogs.groupby('type').sum()
viz['avg_fav'] = viz['favorite_count'] / viz['algo_A_is_dog']

viz
```

```python
WeRateDogs.head()
```

```python
# Visualization
x = ['Not specified', 'doggo', 'floofer', 'pupper', 'puppo']
y = viz.avg_fav
plt.xlabel('Type')
plt.ylabel('Average Favorite count')
plt.title('Each type and it`s average favorite count')
plt.bar(x,y);
```
